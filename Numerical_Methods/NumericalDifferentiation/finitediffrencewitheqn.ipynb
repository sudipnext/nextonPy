```python
# This is for the comparision of the newton three formulas to find the derivative at a point
## Forward Difference 
## BackWard Difference
## Central Difference
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dt = 0.2
t = np.linspace(-2, 4, 60)
def f(x):
    return np.sin(x)
#original function derivative known
def df(x):
    return np.cos(x)
```

```python
plt.plot(t, f(t), linewidth =1.2)
plt.plot(t, df(t), linewidth=3)
plt.legend(["Function", "Exact Derivative"])
plt.show()
```

```python
#Forward difference formula
dfdtF = (f(t+dt)- f(t))/dt 
#Backward
dfdtB = (f(t)-f(t-dt))/dt
#central
dfdtC = (f(t+dt)-f(t-dt))/(2*dt)
```

```python
plt.plot(t, df(t), linewidth=3)
plt.plot(t, dfdtF, '^', linewidth = 1)
plt.plot(t, dfdtB, 'v', linewidth=1)
plt.plot(t, dfdtC, '--', linewidth=1)
plt.legend(["Original","Forward", "Backward", "Central"])
plt.show()
```
