# Finite Difference with different data points

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
#Now we are applying forward difference at initial point and backward at last point and at center we are using central difference formula
n=30
x = np.linspace(0.01, 3, n)

def f(x):
    return np.sin(x)
def df(x):
    return np.cos(x)

dfdx = np.zeros(n)
# for initial applying forward difference
dfdx[0] = (f(x[1])-f(x[0]))/(x[1] - x[0])
for i in range(1, n-1):
    dfdx[i] = (f(x[i+1])-f(x[i-1]))/(x[i+1]-x[i-1])
dfdx[-1] = (f(x[-1]) - f(x[-2]))/(x[-1]-x[-2])
```
