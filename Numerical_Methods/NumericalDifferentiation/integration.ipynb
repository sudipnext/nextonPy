# Numerical Integration Rieman right, left and with Trapezoidal Rule

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Defining the limits
a=0
b=10
dxf = 0.01
xf = np.linspace(a, b, int(b/ dxf))
yf = np.sin(xf)
plt.figure()
plt.grid(True)
plt.plot(xf, yf)

dxc = 0.5
xc = np.linspace(a, b, int(b/dxc))
#End point true means it stopps at last sample
edges = np.linspace(a, b, int(b/dxc)+1, endpoint=True)
yc = np.sin(xc)
plt.stairs(yc, edges, color='r')
plt.grid(True)
plt.show()
```

```python
n= len(xc)

area1=0
#basically using left side value to calculate the area
for i in range(n-1):
    area1 = area1 + (yc[i]*dxc)
#for right side
area2=0
for i in range(n-1):
    area2 = area2 +(yc[i+1]*dxc)
#using trapazoid rule bascially averaging the both
area3=0
for i in range(n-1):
    area3 = area3 + (dxc/2) *(yc[i] + yc[i+1])
print(area1, area2, area3)
```

```python
#using builin functions
area4 = sum(yc[:-1]) *dxc 
area5 = sum(yc[1:]) *dxc
area6 = np.trapz(yc, xc)
print(area4, area5, area6)
```
