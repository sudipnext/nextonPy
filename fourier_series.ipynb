```python
import sympy as sym
from sympy import pi
import numpy as np
```

```python
t= sym.symbols('t')
x = sym.Piecewise((-1, t<0),(1, t>0))
ser = sym.fourier_series(x, (t, -pi, pi))
```

```python
ser.truncate(5)
```

```python
sym.plot(ser.truncate(3), (t, -pi, pi));
```

```python
x1 = sym.Piecewise((-t-2, ((t >=-2)&(t<=-1))),
                   (t, ((t>=-1)&(t<=1))),
                   (-t+2, ((t>=1)& (t<=2))))
```

```python
ser1 = sym.fourier_series(x1, (t, -2, 2))
```

```python
ser1.truncate(4)
```

```python
sym.plot(ser1.truncate(10), (t, -pi, pi))
```

```python
from cmath import exp
for i in np.arange(-20, 20, 1):
    x2 = sym.Piecewise(exp(-i*2*pi*t), (t>=-2)&(t<=2))
```

```python
ser2 = sym.fourier_series(x2, (t, -2, 2))
```

```python
sym.plot(ser2.truncate(10), (t, -pi, pi))
```
