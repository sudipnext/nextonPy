```python
import numpy as np
```

```python
#plotting a sine function 
x=[]
# for j in np.arange(0.0, 10.0, 0.1):
#     x.append(j)
#else we can use linspace
# for i in np.linspace(0, 10, 100):
#     x.append(i)
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

    
```

```python
#plotting exponential
x=[]
for i in np.arange(-8, 8, 0.1):
    x.append(i)


y = np.exp(x)
plt.plot(x, y)
plt.title("Exponential Plotting")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()  
```

```python
#plotting a logarithmic function
x=[]
for k in np.arange(1, 500, 0.1):
    x.append(k)
y = np.log(x)
plt.plot(x, y)
plt.title("Logarithmic function")
plt.show()
```

# Subplots

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y, 'g')
plt.title('Sin')
plt.xlabel('X')
plt.ylabel("sin(y)")
plt.grid()
plt.show()

plt.subplot(2, 1, 2)
plt.plot(x, z, 'r')
plt.title('cos')
plt.xlabel('X')
plt.ylabel("cos(x)")
plt.grid()
plt.show()
```
