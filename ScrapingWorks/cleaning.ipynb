```python
import json
import csv

# Read JSON data from file
with open('6kData.json') as json_file:
    data = json.load(json_file)

# Open a CSV file for writing
with open('newdata.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)

    # Write the header row
    writer.writerow(['name', 'gender', 'occupation', 'height', 'details'])

    # Write each row of data
    for item in data:
        name = item['name']
        gender = item['gender']
        occupation = '{' + ', '.join(json.dumps(occ) for occ in item['occupation']) + '}'
        height = item['height']
        details = json.dumps(item['details'])

        writer.writerow([name, gender, occupation, height, details])
```

### We have the required data in the format i.e name category and subcategory


```python
import re
result=[]
for entry in new_list:
    match = re.match(r'^(.*?) \((.*?)\)$', entry)
    if match:
        name = match.group(1)
        category_subcategory = match.group(2).split(' ')
        if len(category_subcategory) == 1:
            category = None
            subcategory = category_subcategory[0]
        else:
            category = category_subcategory[0]
            subcategory = category_subcategory[1]
        result.append({
            'name': name,
            'category': subcategory,
            'subcategory': category
        })

print(result)
cat_df = pd.DataFrame(result)
cat_df
```
