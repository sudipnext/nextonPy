```python
df = pd.read_csv('fict.csv')
```

```python
df['category'] = 'Fictional'
```

```python
df['name']= df['Name']
df.drop('Name', axis=1, inplace=True)
```

```python
df['height']= df['Height']
df['gender'] = df['Gender']
df.drop('Gender', axis=1, inplace=True)
df.drop('Height', axis=1, inplace=True)
```

```python
df['extras'] = "{'net_worth': None, 'gender': null, 'nationality': None, 'birthday': None, 'age': None, 'is_alive': None, 'death': None, 'birthDate': None, 'deathDate': None}"
```

```python
df = df[['name', 'gender', 'height', 'category', 'subcat1', 'subcat2', 'extras']]
```

```python
df['name'] = df['name'].str.lower()
```

```python
data
```

```python
merged_df = pd.merge(data, df, on='name', how='outer')
```

```python
merged_df
```

```python
import pandas as pd

# Assuming you have already loaded the 'data' and 'df' dataframes and performed the merge

# Fill missing values in "_x" columns with values from "_y" columns
for col in ['gender', 'height', 'category', 'subcat1', 'subcat2', 'extras']:
    merged_df[col + '_x'].fillna(merged_df[col + '_y'], inplace=True)

# Drop the "_y" columns
merged_df.drop(['gender_y', 'height_y', 'category_y', 'subcat1_y', 'subcat2_y', 'extras_y'], axis=1, inplace=True)

# Rename the "_x" columns to remove the suffix
merged_df.rename(columns={'gender_x': 'gender', 'height_x': 'height', 'category_x': 'category', 'subcat1_x': 'subcat1', 'subcat2_x': 'subcat2', 'extras_x': 'extras'}, inplace=True)

# Display the cleaned merged dataframe
print(merged_df)
```

```python
merged_df
```

```python
merged_df.to_csv('merged_df.csv', index=False)
```
