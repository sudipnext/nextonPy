```python
import json
import csv

with open('pokemon.json', encoding='utf-8') as f:
    file_contents = f.read()

data = json.loads(file_contents)

csv_file = 'pokemon.csv'

with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Category', 'Name', 'Height', 'Pro1'])

    for idx, record in enumerate(data):
        if 'documentChange' in record[1][0]:
            fields = record[1][0]['documentChange']['document']['fields']
            if 'list' in fields:
                arrval = fields['list']['arrayValue']['values']
                for i in arrval:
                    findata = i['mapValue']['fields']
                    gen=''
                    cat=''
                    pro=''
                    pro2=''
                    if 'PROP2' in findata:
                        gen = findata['PROP2']
                    if 'stringValue' in gen:
                        gender = findata['PROP2']['stringValue']
                    if 'PROP1' in findata:
                        cat = findata['PROP1']
                    if 'stringValue' in cat:
                        category = findata['PROP1']['stringValue']
                    if 'PROP3' in findata:
                        pro = findata['PROP3']
                    if 'stringValue' in pro:
                        proo = findata['PROP3']['stringValue']
                    # if 'PROP4' in findata:
                    #     pro2 = findata['PROP4']
                    # if 'stringValue' in pro2:
                    #     prooo = findata['PROP4']['stringValue']
                    name = findata['NAME']['stringValue']
                    ht = findata['HEIGHT']
                    if 'doubleValue' in ht:
                        height = ht['doubleValue']
                    else:
                        height = ht['integerValue']

                    csv_writer.writerow([category, name, height, gender, proo])

print(f'Data has been saved to {csv_file}')
```
