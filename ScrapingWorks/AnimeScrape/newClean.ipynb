```python
import pandas as pd
```

```python
df = pd.read_json('animefinal.json')
df.head()
```

```python
df.isnull().sum()
```

```python
urls = pd.read_csv('urls.csv')
urls
```

```python
# Create new columns for "id" and "anime name"
urls['id'] = urls['URL'].str.extract(r'/(\d+)/')
urls['anime name'] = urls['URL'].str.extract(r'/\d+/(.+)/')

# Drop the "URL" column
urls.drop(columns=['URL'], inplace=True)

# Remove hyphens from the "anime name" column
urls['anime name'] = urls['anime name'].str.replace('-', ' ')

```

```python
urls
```

```python
urls.to_csv('processed_anime_data.csv', index=False)
```
