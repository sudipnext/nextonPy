```python
from bs4 import BeautifulSoup
import requests
import json
from multiprocessing import Process, Manager


def scrape_and_save(url, extracted_data):
    try:
        html_data = requests.get(f"https://www.animecharactersdatabase.com/{url}", headers=headers)
        soup = BeautifulSoup(html_data.content, 'html.parser')

        # Find all <tr> elements
        trs = soup.find_all('tr')

        # List to store the extracted data as dictionaries for this URL
        url_data = []

        # Extract the data from the 2nd and 3rd child elements of each <tr>
        for tr in trs:
            tds = tr.find_all('td')
            ths = tr.find_all('th')
            if len(tds) >= 2 and len(ths) >= 1:
                # Get the text from the 2nd child element
                data1 = tds[1].text.strip()  # Gadoras
                # Get the text from the 3rd child element (inside <th>)
                data2 = ths[0].text.strip()  # Height: 188cm
                print(data1, data2)
                # Create a dictionary for each extracted data
                url_data.append({"data1": data1, "data2": data2})

        # Save the extracted data for this URL to the shared list
        extracted_data[url] = url_data

        print(f"Data extracted and saved from {url}")
    except Exception as e:
        print(f"Error while processing {url}: {e}")

if __name__ == "__main__":
    # Use Manager to create a shared dictionary for multiprocessing
    with Manager() as manager:
        # Dictionary to store the extracted data for each URL (separate for each process)
        extracted_data = manager.dict()

        # Create and start the processes
        processes = []
        for url in urls:
            process = Process(target=scrape_and_save, args=(url, extracted_data))
            processes.append(process)
            process.start()

        # Wait for all processes to complete
        for process in processes:
            process.join()

        # Save all the extracted data to a single JSON file
        output_file = 'extr.json'
        with open(output_file, 'w') as json_file:
            json.dump(extracted_data, json_file)

        print("All data extracted and saved to extracted_data.json.")
```
