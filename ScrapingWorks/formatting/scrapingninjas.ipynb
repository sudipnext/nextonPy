```python
import json
import requests
from concurrent.futures import ThreadPoolExecutor
from requests.exceptions import Timeout

# Create an empty list to store the responses
responses = []

def fetch_data(name):
    try:
        api_url = 'https://api.api-ninjas.com/v1/celebrity?name={}'.format(name)
        response = requests.get(api_url, headers={'X-Api-Key': 'dpTwniK5Mx/nvM12bSDzxw==e8MORLclQVJ8dE8f'}, timeout=10)
        
        if response.status_code == requests.codes.ok:
            response_data = response.json()
            if response_data:
                responses.append(response_data[0])
            print(response_data)
        else:
            print("Error:", response.status_code, response.text)
    
    except Timeout:
        print("Connection timeout occurred for", name)

# Create a ThreadPoolExecutor with maximum concurrency of 5
with ThreadPoolExecutor(max_workers=10) as executor:
    # Submit tasks to the executor
    executor.map(fetch_data, test)

# Save the responses to a JSON file
with open('second.json', 'w') as file:
    json.dump(responses, file)
```
