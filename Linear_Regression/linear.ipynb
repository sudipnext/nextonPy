# Linear Regression With Python



```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0, 10.0)

#Reading Data
data = pd.read_csv('C://Users/sudipnext/Downloads/archive/housing.csv')
print(data.shape)
data.head()
```

```python
x= data['median_income'].values
y = data['median_house_value'].values
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
#Calculating X and Y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Total number of values
n = len(x)

#
nume =0
deno = 0
for i in range(n):
    nume += (x[i]-mean_x) * (y[i]- mean_y)
    deno += (x[i]-mean_x) **2
b1 = nume/deno
b0 = mean_y / mean_x

print(b1, b0)
```

```python
from cProfile import label
from turtle import color


max_x = np.max(x)
min_x = np.min(x)

x = np.linspace(min_x, max_x, 1000)
y= b0 + b1 * x
plt.plot(x, y, color='red', label="Regression Line")
plt.scatter(x, y, color='blue', label= 'Scatter Plot')
plt.xlabel('Area')
plt.ylabel("Prices")
plt.show()
```
