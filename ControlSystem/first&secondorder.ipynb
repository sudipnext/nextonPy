```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
from control.matlab import lsim
```

```python
ran = np.linspace(0, 10, 1000)

s= ctl.tf('s')
k=1
TF = k/(s+k)
ran, dom = ctl.step_response(TF, ran)
plt.plot(ran,dom)
plt.xlabel("T")
plt.ylabel("c(t)")
plt.show()
```
