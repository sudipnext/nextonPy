# This is a pratical simulation of how the probability of flipping two coins converges over time to give a value

```python
import matplotlib.pyplot as plt
import numpy as np
```

## A function created to simulate a coin flip

## Let's Check the probability of getting two Tails i.e {T, T} we have four possible outcomes {H,H} {T T} {H T} {T H}
## Theoritically saying we have 0.25 chance of getting each let's see

```python
def coinFlipping():
    coin = ['H', 'T']
    coin1 =['H', 'T']

    coinrandm = np.random.choice(coin)
    coin1randm = np.random.choice(coin1)

    if coinrandm == 'T' and coin1randm == 'T':
        return 1
    else:
        return 0
```

## Let's simulate the no.of trials as the no.of trials increases the probabilty of getting true value increases

```python
noflips = int(input())
counting_tails = 0
array_of_TnTforeachflip =[]
array_of_no_of_times =[]
for eachflip in range(noflips):
    counting_tails += coinFlipping()
    array_of_TnTforeachflip.append(counting_tails/(eachflip+1))
    array_of_no_of_times.append(eachflip+1)
```

## simulation on matplotlib

```python
from cProfile import label
from turtle import color


plt.plot(array_of_no_of_times, array_of_TnTforeachflip, label="Experemental findings")
plt.title('Simulation of Coin Flipping over time')
plt.xlabel('No. of times=>')
plt.ylabel('proportion of comming T and T for each trail increasing')
plt.axhline(y = 0.25, color = 'r', label= "True Probability", linestyle = '-')
plt.legend()
plt.show()
```
